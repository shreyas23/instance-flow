```python
## create car/pedestrian tracking dataset

import os
import glob
import numpy as np
import skimage.io as sio
import matplotlib.pyplot as plt
from multiprocessing import Pool

gt_root = os.path.join("/external/datasets/kitti/tracking", "kitti-step", "panoptic_maps", "original", "val")
save_root = os.path.join("/external/datasets/kitti/tracking", "kitti-step", "panoptic_maps", "trimmed", "val")

gt_files = glob.glob(os.path.join(gt_root, "*/*.png"))
save_files = [x.replace("val", "val_mod") for x in gt_files]

args = zip(gt_files, save_files)

def modify_gt(args):
    file_path, save_path = args
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    gt = sio.imread(os.path.join(gt_root, file_path))
    gt[gt[:, :, 0] < 11] = np.array([0, 0, 0])
    gt[gt[:, :, 0] == 255] = np.array([0, 0, 0])
    sio.imsave(save_path, gt)

with Pool() as p:
    p.map(modify_gt, args)
```
